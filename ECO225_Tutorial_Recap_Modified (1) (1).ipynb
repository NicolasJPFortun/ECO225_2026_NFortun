## Git Troubleshooting

**Don't panic about Git.** For this course, you only need these 5 commands:

### The Essential Git Workflow:

```bash
# 1. CHECK STATUS (what's changed?)
git status

# 2. ADD FILES (stage your changes)
git add filename.ipynb          # Add one file
git add .                       # Add everything

# 3. COMMIT (save a checkpoint)
git commit -m "Completed Week 2 tutorial"

# 4. PUSH (upload to GitHub)
git push origin main

# 5. PULL (download latest from GitHub - before starting work)
git pull origin main
```

### Common Git Errors & Fixes:

#### Error 1: "Authentication failed"
**Fix:** Use Personal Access Token (not your password)
- GitHub.com → Settings → Developer Settings → Personal Access Tokens → Generate new token
- Use this token as your password

#### Error 2: "Please commit your changes or stash them"
**Fix:** 
```bash
git stash          # Save your work temporarily
git pull           # Get updates
git stash pop      # Restore your work
```

#### Error 3: "Fatal: not a git repository"
**Fix:** You're in the wrong folder
```bash
cd path/to/your/repo    # Navigate to your repo first
```

#### Error 4: Committed the wrong thing
**Fix:** Undo last commit (keeps your changes)
```bash
git reset --soft HEAD~1
```

**That's it. You don't need to know more than this for ECO225.**

## Exploratory Data Analysis - Complete Review

EDA is about **understanding your data** before formal analysis. Three main components:

### 1. Univariate Analysis (One variable at a time)

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
df = pd.DataFrame({
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'GDP_Growth': [2.6, 2.1, 1.4, 0.8, 1.2],
    'Unemployment': [3.8, 5.4, 4.2, 3.1, 7.2],
    'Inflation': [3.2, 3.9, 4.7, 2.8, 2.3]
})

# Descriptive statistics
print("Summary Statistics:")
print(df['Unemployment'].describe())

# Histogram - distribution of one variable
plt.figure(figsize=(8, 5))
plt.hist(df['Unemployment'], bins=5, edgecolor='black', color='skyblue')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Unemployment Rates')
plt.show()
````

### 2. Bivariate Analysis (Relationship between two variables)

````python
# Correlation
correlation = df[['Unemployment', 'Inflation']].corr()
print("Correlation Matrix:")
print(correlation)
print(f"\nCorrelation coefficient: {correlation.iloc[0, 1]:.3f}")

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Unemployment', y='Inflation', s=100)
sns.regplot(data=df, x='Unemployment', y='Inflation', scatter=False, color='red')
plt.title('Unemployment vs Inflation')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Inflation Rate (%)')
plt.grid(True, alpha=0.3)
plt.show()
````

### 3. Grouping and Comparison

````python
# Create categories
df['GDP_Category'] = pd.cut(df['GDP_Growth'], 
                             bins=[0, 1.5, 2.5, 10],
                             labels=['Low', 'Moderate', 'High'])

# Box plot - compare distributions across groups
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='GDP_Category', y='Unemployment', hue='GDP_Category', legend=False)
plt.title('Unemployment by GDP Growth Category')
plt.xlabel('GDP Growth Category')
plt.ylabel('Unemployment Rate (%)')
plt.show()
````

### Key EDA Concepts to Remember:

**Correlation:**
- Range: -1 to +1
- Positive: variables move together
- Negative: variables move in opposite directions
- Zero: no linear relationship

**Extracting correlation:**
```python
# Method 1: .iloc (position-based)
corr_value = correlation_matrix.iloc[0, 1]

# Method 2: .loc (label-based)
corr_value = correlation_matrix.loc['Unemployment', 'Inflation']
```

**Visualization Types:**
- **Histogram:** Distribution of one variable
- **Scatter plot:** Relationship between two continuous variables
- **Box plot:** Compare distributions across categories
- **Heatmap:** Visualize correlation matrix

## Pandas Cheat Sheet

Focus on these core operations:

### Selection & Filtering:

````python
# Select ONE column → Series
df['Country']

# Select MULTIPLE columns → DataFrame
df[['Country', 'GDP_Growth']]

# Filter rows by condition
df[df['Unemployment'] > 4.0]

# Multiple conditions (use & for AND, | for OR)
df[(df['Unemployment'] > 4.0) & (df['Inflation'] < 4.0)]

# Access by position
df.iloc[0]        # First row
df.iloc[0, 1]     # First row, second column

# Access by label
df.loc[0, 'Country']
````

### Creating & Modifying Columns:

````python
# Add new column
df['Misery_Index'] = df['Unemployment'] + df['Inflation']

# Conditional column with np.where
df['Inflation_Status'] = np.where(df['Inflation'] >= 4.0, 'High', 'Low')

# Categorical column
df['GDP_Category'] = pd.cut(df['GDP_Growth'], 
                             bins=[0, 1.5, 2.5, 10],
                             labels=['Low', 'Moderate', 'High'])
````

### Sorting & Grouping:

````python
# Sort by one column
df.sort_values('GDP_Growth', ascending=False)

# Sort by multiple columns
df.sort_values(['GDP_Growth', 'Inflation'], ascending=[False, True])

# Group by and aggregate
df.groupby('GDP_Category')['Unemployment'].mean()

# Multiple aggregations
df.groupby('GDP_Category')[['Unemployment', 'Inflation']].agg(['mean', 'std'])
````

### The Most Common Mistake:

**WRONG:**
```python
df['Country', 'GDP_Growth']  # Missing extra brackets!

```
**CORRECT:**
```python
df[['Country', 'GDP_Growth']]  # Double brackets for multiple columns
```

**Why?** 
- Single column: `df['Country']` → one set of brackets
- Multiple columns: `df[['Country', 'GDP_Growth']]` → list inside brackets
